```python


match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df["current_score"]
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
# Replace missing dismissal info with "0"
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

# Convert player names to "1" (if out) and keep "0" as it is
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: "0" if x == "0" else "1")

# Change data type to integer
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

# Cumulative sum of dismissals grouped by match
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values

# Calculate remaining wickets
delivery_df['wickets'] = 10 - wickets

# Check updated data
delivery_df.head()
```

```python
delivery_df.head()
```

```python
# Calculate Current Run Rate (CRR) at each delivery
# Formula: (current_score * 6) / (balls played so far)
delivery_df['crr'] = (delivery_df['current_score'] * 6) / (120 - delivery_df['balls_left'])

```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
import numpy as np

# Calculate Required Run Rate (RRR)
delivery_df['rrr'] = (delivery_df['runs_left'] * 6) / delivery_df['balls_left']

# Handle division by zero by replacing infinite values with zero
delivery_df['rrr'] = delivery_df['rrr'].replace([np.inf, -np.inf], 0)

# Now select the required columns into the final DataFrame
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left',
                        'wickets', 'total_runs_x', 'crr', 'rrr', 'result']]

```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Define the ColumnTransformer
trf = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'  # Keep the other columns as they are
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])

```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import numpy as np

# Match summary function
def match_summary(row):
    print(f"Batting Team: {row['batting_team']} | Bowling Team: {row['bowling_team']} | Target: {row['total_runs_x']}")

# Match progression function
def match_progression(x_df, match_id, pipe):
    # Filter the data for the specific match
    match = x_df[x_df['match_id'] == match_id]
    
    # Check if any data is available for the given match_id
    if match.empty:
        print(f"No data found for match_id {match_id}")
        return None, None
    
    match = match[match['ball'] == 6]  # Select rows where the ball count is 6 (end of each over)
    
    # Check if the filtered data for overs is empty
    if match.empty:
        print(f"No overs data found for match_id {match_id}")
        return None, None
    
    # Prepare the DataFrame with the required features
    temp_df = match[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]  # Drop rows where balls_left is 0

    # Check if the DataFrame after filtering is empty
    if temp_df.empty:
        print(f"No valid data available after filtering for match_id {match_id}")
        return None, None

    # Predict win/loss probabilities using the trained model
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0] * 100, 1)  # Probability of losing
    temp_df['win'] = np.round(result.T[1] * 100, 1)   # Probability of winning
    temp_df['end_of_over'] = range(1, temp_df.shape[0] + 1)  # Number the overs
    
    # Get target runs and calculate additional statistics
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    
    # Insert the target value as the first item and calculate runs scored after each over
    new_runs = runs[:]
    runs.insert(0, target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    
    # Calculate the number of wickets lost in each over
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0, 10)  # Assume the team starts with 10 wickets remaining
    wickets.append(0)  # Append 0 as placeholder for the last row
    w = np.array(wickets)
    nw = np.array(new_wickets)
    
    # Calculate the number of wickets fallen in each over
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]

    # Print the target score
    print(f"Target: {target}")

    # Return the resulting DataFrame with match progression details
    temp_df = temp_df[['end_of_over', 'runs_after_over', 'wickets_in_over', 'lose', 'win']]
    return temp_df, target

# Example usage:
# Assuming 'delivery_df' is your dataframe and 'pipe' is the trained model.
# Example: Call the function with match ID 74
temp_df, target = match_progression(delivery_df, 74, pipe)

# Check if data is returned before printing
if temp_df is not None:
    print(temp_df)

    
```
